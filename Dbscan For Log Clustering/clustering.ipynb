Download logs

```python
import requests

url = "https://raw.githubusercontent.com/logpai/loghub/master/HealthApp/HealthApp_2k.log"
response = requests.get(url)

with open('data/HealthApp_2k.log', 'wb') as file:
    file.write(response.content)
```

```python
import pandas as pd

# Read the log file
log_data = pd.read_csv('data/HealthApp_2k.log', delimiter='\t', header=None)

# Display the first few rows of the log data
print(log_data.head())

# Display basic information about the log data
print(log_data.info())

# Display basic statistics of the log data
print(log_data.describe())
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import cosine_distances

# Load the pre-trained sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for the log data
embeddings = model.encode(log_data[0].tolist())
```

```python

# Perform DBSCAN clustering
dbscan = DBSCAN(metric='cosine', eps=0.1, min_samples=1)
clusters = dbscan.fit_predict(embeddings)

# Add the cluster labels to the log data
log_data['Cluster'] = clusters

# Display the first few rows of the log data with cluster labels
print(log_data.head(20))
```

```python
log_data[log_data.Cluster == 0].head(10)
```

```python
# Count the number of records in each cluster
cluster_counts = log_data['Cluster'].value_counts()

# Filter clusters that have more than 10 records
clusters_with_more_than_10_records = cluster_counts[cluster_counts > 10].index

# Print 5 log_data from each of these clusters
for cluster in clusters_with_more_than_10_records:
    print(f"Cluster {cluster} (Total records: {cluster_counts[cluster]})")
    print(log_data[log_data['Cluster'] == cluster].head(5))
    print("\n")
```

```python
# Count the number of records in each cluster
cluster_counts = log_data['Cluster'].value_counts()

# Sort the log_data by cluster
sorted_log_data = log_data.sort_values(by='Cluster')

# Write the sorted log_data to a CSV file
sorted_log_data.to_csv('sorted_log_data.csv', index=False)
```
